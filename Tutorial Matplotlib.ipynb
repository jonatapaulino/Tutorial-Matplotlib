```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# para visualização direta no jupyter notebook (não é necessário em outros IDE´s)
%matplotlib inline

# caso você esteja usando outra IDE, é fundamental que sua última linha de código
# seja->  plt.show()           Caso contrário, as imagens não serão apresentadas! 
```

Aqui produzimos alguns dados para posterior plotagem:

```python
reta = np.arange(1,21)
aleat = np.random.randint(1,20,20)
curva = [(i/4)**2 for i in reta]
```

A forma mais simples de plotar dados é usando o plot() e passando os dados desejados como argumento

```python
plt.plot(reta)
```

<b>Aqui já começamos a personalizar nosso gráfico, nesse exemplo podemos usar dois argumentos:<br>
*    color = 'red' ou color = 'r'  =>  cor da linha, no caso escolhemos vermelho      <br>
*    linestyle= '--'  ou ls = '--'  => estilo da linha    -- significa  tracejado     <br>

Ou resumir tudo isso, usando apenas 'r--'

```python
# plt.plot(aleat, color = 'red', ls = '--')  essa linha e a de baixo produziriam o mesmo resultado
plt.plot(aleat, 'r--')

plt.xlabel('eixo X')
plt.ylabel('eixo Y')
plt.title('EXEMPLO')
```

<b>Outros estilos de linha (linestyle ou ls):
*    'o'  =>  bolas
*    '.' => pontos
*    '--' => tracejado
*    '-.' => traços e pontos
*    '\*' => estrelas

```python
# Para colocar mais de um curva nos mesmos eixos, basta plotá-las em sequência: 
plt.plot(curva, 'yo',  label ='curva')    # bolas amarelas (y de yellow)
plt.plot(aleat, 'g-.', label = 'aleat')   # traços e pontos verdes (green)
plt.plot(reta, 'b*',   label = 'reta')    # estrelas pretas
plt.title('EXEMPLO')                      # Define o título dos eixos plotados 

plt.legend()                              # Obrigatório se quiser mostrar as legendas!
```

```python
# Para acrescentar outros eixos na mesma figura, é só usar a função subplot(), 
# passando o número de linhas, o n de colunas e o eixo que se deseja plotar. Exemplo:

plt.subplot(1, 2, 1)    # 2 linhas, 1 coluna, plotar no eixo 1
plt.plot(curva, 'y.')   # plota esses dados no primeiro eixo

plt.subplot(1, 2, 2)    # 2 linhas, 1 coluna, plotar no eixo 2
plt.plot(aleat, 'r--')  # plota esses dados no segundo eixo
```

```python
# Outro exemplo:

for i in range (1, 10):
    plt.subplot(3,3,i)       # nº linhas , nº colunas, nº plot
    plt.plot(aleat, 'r-')    # plota o mesmo gráfico 9 vezes
    
plt.tight_layout()           # para evitar sobreposição de imagens, usamos o tight_layout() que organiza tudo automaticamente
```

<br>
<b>Outra forma (talvez a melhor) de se plotar diversos gráficos é usando Programação Orientada a Objetos </b><br>
Essa forma pode parecer mais complexa no início, mas quando você aprende, te dá mais possibilidades

```python
# Instanciamos um objeto usando o figure()
fig = plt.figure()


# Em seguida adicionamos eixos a esse objeto com add_axes(), passando uma lista com as coordenadas de tamanho e localização

eixos1 = fig.add_axes([0,0,1,1])        # São quatro valores de 0 - 1: começo do eixo x, começo do eixo y, largura e altura
eixos1.plot(curva, 'g*', label='curva')
eixos1.set_title('QUADRO GRANDE')       
eixos1.legend(loc=4)                    # loc define a localização da legenda


# Agora vamos adicionar outros eixos de forma que tenhamos um quadro pequeno dentro do grande que já foi plotado

eixos2 = fig.add_axes([0.1, 0.5,.4,.4])    # É aqui que escolhemos localização e tamanho (x, y, largura e altura)
eixos2.plot(aleat, 'r--', label='aleat')
eixos2.set_title('quadro pequeno')
eixos2.legend(loc=4)                       # Define a localização da legenda, passar um int de 0 a 10
```

<b>Criando vários gráficos na mesma figura com OOP

Perceba que usaremos usando tuple unpacking - "desempacotamento de tupla" na função subplots, onde: <br>
O primeiro item é a figura em si (ex: fig) e; <br>
O segundo é um objeto (ex: eixos) com o qual podemos iterar por todos os eixos criados

```python
# Subplot() => Parâmetros obrigatórios: nº de linhas e colunas - nrows e ncols ; 
# Parâmetros opcionais: tamanho da figura - 'figsize' recebe uma tupla;  resolução 'dpi' 

fig, eixos = plt.subplots(nrows=2, ncols=1, figsize=(8,6), dpi=70) 
eixos[0].plot(reta, label = 'reta', lw=4)
eixos[0].set_title("Exemplo 1")
eixos[0].set_yticks([2, 6, 15, 19, 25])  # determina ticks para o eixo selecionado 

eixos[1].plot(curva, '--', label = 'curva')
eixos[1].set_title("Exemplo 2")
eixos[1].set_yscale('log')       # adiciona escala logaritmica

for eixo in eixos:
    eixo.legend()
    
plt.tight_layout()    # impede sobreposição
```

Outro exemplo com mais opções de personalização: <br>
Veja que podemos personalizar tanto a linha, quanto o marcador (marker)    

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(curva, color='red', label = 'curva', ls= '-.')                    # ls - estilo da linha
ax.plot(reta, label = 'reta', lw=5, alpha=0.5,                            # lw - grossura da linha; alpha - transparência;
        marker = '*', ms = 20, mfc = 'yellow', mew = 2,  mec = 'purple')  
# marker= marcador dos pontos; ms= tamanho; mfc= cor, mew= calibre da borda, mec= cor da borda

ax.legend(loc=4)
ax.set_xlim(1,20)      # escolhe limite de visualização para o eixo x
ax.set_ylim(1,18)      # escolhe limite de visualização para o eixo y
```

<b>Existem outros tipos de gráficos, como dispersão, barras, degrau, preenchido e outros<br>
Para usá-los basta substituir o 'plot' que estávamos usando, pela respectiva função:

```python
print('\n OUTROS EXEMPLOS: \n')
n = np.array([1,2,3,4,5,6,7])

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10,4), dpi = 200)

axes[0].scatter(aleat, aleat + 30*np.random.randn(len(aleat)), color= 'black', alpha = .3)
axes[0].set_title("Dispersão")

axes[1].bar(n, n**2, color='green')
axes[1].set_title("Barras")

axes[2].step(n, n**2, 'red', linestyle = '--')
axes[2].set_title("Degrau")

axes[3].fill_between(n, n**2, n**3, color="y", alpha = 0.6);
axes[3].set_title("Preenchido");

plt.tight_layout()
```
